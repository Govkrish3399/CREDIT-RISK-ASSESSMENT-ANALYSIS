##  Merged the two datasets like application and previous_application using Outer join function in python

```python
merg_data = pd.merge(credit_data, previous_data, on='SK_ID_CURR', how='outer' )
```

```python
merg_data
```

```python
merg_data.shape
```

## Handling Missing Values in the Merged Banking data

```python
merg_data.isnull().sum()
```

## DROPPING COLUMNS WITH MISSING VALUES

```python
columns_with_missing_values = ['FLAG_OWN_REALTY','FLAG_OWN_CAR','CNT_CHILDREN','DAYS_FIRST_DUE','DAYS_LAST_DUE']
```

```python
data_cleaned = merg_data.dropna(subset=columns_with_missing_values)
print(data_cleaned)
data_cleaned.shape
```

Checking the data imbalance by analyzing the distribution of target variable which is credit decision

```python
data_cleaned.TARGET.value_counts(normalize=True)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=data_cleaned, x='TARGET')
plt.title('Credit Decision Count')
plt.xlabel('Credit Decision')
plt.ylabel('Count')
plt.show()
```

The above distribution plot shows a imbalance in the target variable. This imbalance obsereved could affect the model performance.

## Lets address Business questions by performing EDA on the merged bank data

 Q.1 What all features are correlated with target variable credit decision?

```python
correlation_matrix = data_cleaned.corr()
print(correlation_matrix)

```

Q.2 What would be the top 20 correlated features with the target variable credit decision?

```python
top_correlations =correlation_matrix['TARGET'].abs().sort_values(ascending=False).head(21)
print(top_correlations)

```

Q.3 What would be top 3 most correlated features with the target variable?

```python
sns.heatmap(data_cleaned[['EXT_SOURCE_3','EXT_SOURCE_1','EXT_SOURCE_2','DAYS_BIRTH','TARGET']].corr(), annot=True, cmap='coolwarm')
```

From the above correlations, it could be noticed that EXT_SOURCE_1, EXT_SOURCE_2 and EXT_SOURCE_3 variables are showing negative correlation with the target variable.These variable often represent credit score which measures the customer credit worthiness during loan decision. Another interesting thing to notice here is DAYS_BIRTH or age is showing a positive correlation with the target, which could indicate age is a important feature to be considered in the credit decision or loan approval,since it measures one's financial stability, repayment capacity.

Q.4 How is education related to Credit decision?

```python
plt.figure(figsize=(10, 6))
edu_decision = data_cleaned.groupby('NAME_EDUCATION_TYPE')['TARGET'].value_counts(normalize=True).unstack()
edu_decision.plot(kind='bar', stacked=True)
plt.title('Credit Decision Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Percentage')
plt.legend(title='Credit Decision')
plt.show()
```

Inferences: Stacked bar shows the likelihood of credit approval based on educational catergories. Its interesting to notice that, customers with higher education level will get loan approved easily.O indicates all other cases and 1 indicates client with payment difficulties.

Q.5 How would the contract type influence loan approval?

```python
# Calculate default rates by contract type
contract_default_rates = data_cleaned.groupby("NAME_CONTRACT_TYPE_x")["TARGET"].mean().sort_values(ascending=False)
print(contract_default_rates)
```

```python
plt.figure(figsize=(6, 4))
contract_default_rates.plot.bar(color=['Red','Yellow'])
plt.title("Default Rates by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Default Rate")
#plt.ylim(0, 0.2) sets the lower limit of the y-axis to 0 and the upper limit to 0.2
plt.ylim(0,0.2)
plt.show()
```

The plot shows the default rates by contract type. Its interesting to notice here that Cash loan defaut rate is greater than Revolving loans. The bar plot indiciates the differences in default rates of different contract types.
